```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()

# Add an Axes object to the figure
ax = plt.axes(projection='3d')

# Generate random data
xdata = np.random.standard_normal(100)
ydata = np.random.standard_normal(100)
zdata = np.random.standard_normal(100)

# Create a 3D scatter plot
ax.scatter3D(xdata, ydata, zdata)

# Display the plot
plt.show()
```

```python
#Scatter Plot of "Rating" and "Revenue (Millions)"

# Create a scatter plot matrix
df.plot.scatter(x='Rating', y='Revenue (Millions)')
plt.show()
```

```python
#Scatter Plot of "Metascore" and "Votes"
df.plot.scatter(x='Metascore', y='Votes')
plt.show()
```

```python
#kde-Plot
import seaborn as sns
geyser = sns.load_dataset("geyser")
sns.kdeplot(data=geyser, x="waiting", y="duration")
```

```python
#Create a Histplot
#Reviewing the Rank of the movies
df.hist(column='Rank')
```
