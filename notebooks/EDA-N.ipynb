## Univariate Analysis 

# Task: Determine the distributions of each numeric column and gauge any outliers 
# Common graphs for univariate analysis: histograms

```python
# histogram for the 'Rank' column

sns.histplot(data=df, x='Rank')
```

This seems to be an example of a column where univariate analysis would not be helpful with as the rankings are in chronological order and each movie has their own rank. Causing the distribution to be stagnant.

```python
# histogram for the 'Year' column 

sns.histplot(data=df, x='Year')
```

This can be seen as an 'left skewed' distribution which tells us that amongst the top 1000 movies in IMDB from 2006-2016, most of them fall in the year of 2016 or in general, in the later years. 

```python
# histogram for the 'Runtime' column

sns.histplot(data=df, x='Runtime (Minutes)')
```

Potential outliers may be movies that range in the 70 minutes range and movies that are 180 minutes and above. The distribution can be seen as 'right skewed' meaning that the popular movies listed in IMDB typically go for shorter run times. 

```python
# histogram for the 'Rating' column 

sns.histplot(data=df, x='Rating')
```

This graph showcases a 'left skewed' distribution which means to tell us that typically the rating for the most popular movies on IMDB are on the higher range with positive reviews possibly. 

```python
# histogram for the 'Votes' column 

sns.histplot(data=df, x='Votes')
```

We notice a 'right skewed' distribution in this histogram, meaning that voting for determining the most popular movies in this dataset are on the lower end. However, it can be seen that there is an outlier of 1.75 million for one movie.

```python
# histogram for the 'Revenue' column 

sns.histplot(data=df, x='Revenue (Millions)')
```

This histogram shows a 'right skewed' distribution meaning that the most popular movies within this dataset typically surprisinly did not make high amounts of income. There is also notably an outlier with a movie grossing about 900 million in revenue. 

```python
# histogram for the 'Metascore' column 

sns.histplot(data=df, x='Metascore')
```

Metascores refer to the rating of a film given by the 'world's most respected critics.' This histogram showcases a 'normal' distribution, meaning that the reviews given by critics tend to cluster around the mean of the distribution which is about 60. 

## Bivariate Analysis 

# Now that we have gauged the distributions of each of the numeric columns in this dataset. Let's try and see if there are any relationships between any of the numeric columns.

```python
# boxplot to showcase possible relationship between 'Rating' and 'Runtime' columns 

sns.boxplot(data= df, x= 'Rating', y= 'Runtime (Minutes)')
```

This seems hard to read. Let's try adjusting the plot size.  

```python
plt.figure(figsize=(20, 14))
# 20 = width (in) and 14 = height (in)
sns.boxplot(data= df, x= 'Rating', y= 'Runtime (Minutes)')
```

Since there are so many boxplots within this graph, we should prioritize certain aspects of the plots. Looking at the line inside each of the plots (median) for the higher ratings, we can see a pattern that they average to be for movies with about 120 minutes in runtime. However, it is important to note the spread of the boxplots in the higher ratings showcasing a high variability amongst the data, leading to uncertainty in making generalizations of the data.  

```python
df.plot(kind='scatter', x='Runtime (Minutes)', y='Rating', title='Runtime Vs. Rating')
plt.show()
```

This scatter plot reinforces the notion that higher ratings for the most popular movies are most notably within the 100-120 minutes range.
